### ****Importación de datos****



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Sumatoria del total de ingresos de cada tienda basados en los precios
ingreso_t1 = tienda['Precio'].sum()
ingreso_t2 = tienda2['Precio'].sum()
ingreso_t3 = tienda3['Precio'].sum()
ingreso_t4 = tienda4['Precio'].sum()
```

```python
#Visualizar los ingresos totales por tienda en formato moneda
def formato_moneda(valor):
    return f"${valor:,.2f}"

print("Ingresos totales de la Tienda 1:", formato_moneda(ingreso_t1))
print("Ingresos totales de la Tienda 2:", formato_moneda(ingreso_t2))
print("Ingresos totales de la Tienda 3:", formato_moneda(ingreso_t3))
print("Ingresos totales de la Tienda 4:", formato_moneda(ingreso_t4))
```

# 2. Ventas por categoría

```python
#Agrupar y ordenar las categorias por cada tienda
ventas_categoria_t1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_t2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_t3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_categoria_t4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
```

```python
#Imprimir por cada tienda las categorias
print("Tienda 1")
print(ventas_categoria_t1)
print("\nTienda 2")
print(ventas_categoria_t2)
print("\nTienda 3")
print(ventas_categoria_t3)
print("\nTienda 4")
print(ventas_categoria_t4)
```

```python
#Mostrar en tablas ordenadas las categoras por cada tienda para mejor visualización
df_t1 = ventas_categoria_t1.reset_index()
df_t1.columns = ['Categoría del Producto', 'Cantidad_Ventas']

df_t1
```

# 3. Calificación promedio de la tienda


```python
#Calcular promedios de calificacion por cada tienda
prom_t1 = tienda['Calificación'].mean()
prom_t2 = tienda2['Calificación'].mean()
prom_t3 = tienda3['Calificación'].mean()
prom_t4 = tienda4['Calificación'].mean()
```

```python
#Mostrar promedio de calificación de usuarios por cada tienda
calificaciones = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Promedio Calificacion': [prom_t1, prom_t2, prom_t3, prom_t4]
})

calificaciones['Promedio Calificacion'] = calificaciones['Promedio Calificacion'].round(2)

calificaciones
```

# 4. Productos más y menos vendidos

# 5. Envío promedio por tienda

```python
#Calcular promedio envio por tienda
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()
```

```python
#Tabla ordenada con el costo de envio promedio por tienda
envios = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo Envío Promedio': [envio_t1, envio_t2, envio_t3, envio_t4]
})

envios['Costo Envío Promedio'] = envios['Costo Envío Promedio'].apply(formato_moneda)

envios
```
